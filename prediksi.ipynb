```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("Indeks Standar Pencemar Udara di Provinsi DKI Jakarta 2021.csv")
df
```

```python
df.isnull().sum()
```

```python
df['pm25'].fillna(value=df['pm25'].mean(), inplace=True)
```

```python
df['tanggal']= pd.to_datetime(df['tanggal'])
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

print(df['critical'].unique(), '\n')
print(df['categori'].unique(), '\n')
labelencoder = LabelEncoder()
df['critical'] = labelencoder.fit_transform(df['critical'])
print(df['critical'].unique(), '\n')
df['categori'] = labelencoder.fit_transform(df['categori'])
print(df['categori'].unique(), '\n')
```

```python
df
```

```python
X = df[['pm10', 'pm25', 'so2', 'co', 'o3']]
y = df['categori']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)

print('MAE : ', mean_absolute_error(y_test, y_pred))
print('MSE : ', mean_squared_error(y_test, y_pred))
print('R2 : ', r2_score(y_test, y_pred))
```

```python
future_date = [[86, 58, 24, 64, 47]]
future_pred = rf_model.predict(future_date)
print('Prediksi Indeks Pencemaran Udara : ', future_pred)
```
